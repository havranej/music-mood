```python
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import ShortTermFeatures

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm import tqdm

%matplotlib inline
```

```python
fs, x = audioBasicIO.read_audio_file("../data/clips_45seconds/2.mp3")
```

```python
WINDOW = 0.020
STEP = 0.010
features, f_names = ShortTermFeatures.feature_extraction(x, fs, WINDOW*fs, STEP*fs) # Default: Window width 50 ms, step 25 ms
```

```python
features.shape
```

```python
df = pd.DataFrame(features.T, columns = f_names)
df["time"] = [x * STEP for x in range(features.shape[1])]
df["song_id"] = 2
df
```

```python
WINDOW = 0.020
STEP = 0.010

dfs = []

for i in tqdm(range(2, 1001)):
    fs, x = audioBasicIO.read_audio_file(f"../data/clips_45seconds/{i}.mp3")
    features, f_names = ShortTermFeatures.feature_extraction(x, fs, WINDOW*fs, STEP*fs)
    df = pd.DataFrame(features.T, columns = f_names)
    df["time"] = [x * STEP for x in range(features.shape[1])]
    df["song_id"] = i
    dfs.append(df)
    df.to_csv(f"../data/paa_features_dense/{i}.csv", index = False, float_format='%.6f')
```
